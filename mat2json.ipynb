```python
import h5py
import matplotlib.pyplot as plt
import numpy as np

# Open the file
f = h5py.File('/Volumes/scratch/ophys/BCI/active_mice/733585_SBCI24/SLAP2/slap2_733586_2024-05-20_13-30-42/Neuron1/BCIon_20240520_142619_DMD2.meta', 'r')

keys = []
values = []

# Define a function to be called on each object in the file
def print_attrs(name, obj):
    print(name)
    for key, val in obj.attrs.items():
        print("    %s: %s" % (key, val))
        keys.append(key)
        if isinstance(val, np.ndarray) and val.size == 1:
            values.append(np.asscalar(val))
        else:
            values.append(val)

# Call the visititems method with the function
f.visititems(print_attrs)

# Plot the keys and values
plt.bar(keys, values)
plt.show()
```

# Get mouse ID

```python
import os
import glob
import re
from datetime import datetime, timezone

def get_mouse_info(base_path):
    for root, dirs, files in os.walk(base_path):
        # Check if we're in a Neuron* directory
        if os.path.basename(root).startswith('Neuron'):
            # Look for BCIon_* files
            bci_files = glob.glob(os.path.join(root, 'BCIon_*'))
            if bci_files:
                # Extract mouse ID and date-time from the path
                path_parts = root.split(os.sep)
                mouse_id = None
                date_time = None
                for part in path_parts:
                    if '_SBCI' in part:
                        mouse_id = part
                    if re.match(r'slap2_\d+_\d{4}-\d{2}-\d{2}_\d{2}-\d{2}-\d{2}', part):
                        date_time = part.split('_', 2)[-1]
                if mouse_id and date_time:
                    return mouse_id, date_time
    
    # If no matching files found
    return None, None

# Example usage
base_path = '/Volumes/scratch/ophys/BCI/active_mice'
mouse_id, date_time = get_mouse_info(base_path)

if mouse_id and date_time:
    print(f"Mouse ID: {mouse_id}")
    print(f"Date-Time: {date_time}")
else:
    print("No matching mouse information found")

date_string = date_time.replace("_", " ")  # replace underscore with space
date_string = date_string.replace("_", " ")  # replace underscore with space

# Parse the string into a datetime object
# dt = datetime.strptime(date_string, "%Y-%m-%d %H-%M-%S")

# Set the timezone to UTC
# dt = dt.replace(tzinfo=timezone.utc)

print(date_time)  # Output: 2024-05-20 13:30:42+00:00
```
